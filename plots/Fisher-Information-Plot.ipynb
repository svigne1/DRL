```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python

x = np.linspace(0.0001, 1000, 100000)

fig = plt.figure()
plt.plot(x, np.log(x), '-')
```

```python
mu = 0
sigma = 5
x = np.arange(-100, 100, 0.001)
p1 = norm.pdf(x,mu,sigma)
logp1 = norm.logpdf(x,mu,sigma)
p2 = norm.pdf(x,mu,sigma+25)
logp2 = norm.logpdf(x,mu,sigma+25)

fig, axs = plt.subplots(2,2,)
axs[0,0].plot(x, p1)
axs[1,0].plot(x, logp1)
axs[0,1].plot(x, p2)
axs[1,1].plot(x, logp2)
```
